```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings 

warnings.filterwarnings('ignore')
```

```python
mysales_electronics =pd.read_excel("eletronic_sales.xlsx")
print( 'Teble Head')
mysales_electronics.head(44)
```

```python
mysales_electronics.shape   # this tells us the number of row is 43 and columns 5
```

```python
# Give us insights into the Revenue is calculated
mysales_electronics["Revenue"] = mysales_electronics["Units"] * mysales_electronics["Price"]
mysales_electronics["Revenue"]
```

```python
# round(parameter,2)
mysales_electronics["Revenue(N)"] = round(mysales_electronics["Units"]*mysales_electronics["Price"] * 1000,2)
mysales_electronics["Revenue(N)"]
```

```python

mysales_electronics.dtypes.value_counts().plot(kind="bar", xlabel = "Data Types", ylabel = "Count")
```

```python
# Getting all the total sales done on electronics sales

Total_goods_sold = mysales_electronics["Units"].sum()
print(f"Total goods sold: {Total_goods_sold} units")
```

```python
# Now we need to know Total Revenue
Total_Revenue = mysales_electronics["Revenue(N)"].sum()
print(f"Total Revenue: N{Total_Revenue}")
```

```python
#using "round() function to approximate to 2 decimal place"
Average_revenue_per_unit = round(Total_Revenue / Total_goods_sold,2)
print(f"Average Revenue per unit: N{Average_revenue_per_unit}")
```

```python
# mysales_electronics = mysales_electronics.drop(columns=['Month_Name'])
mysales_electronics['Year'] = mysales_electronics['Date'].dt.year
mysales_electronics['Month_Name'] = mysales_electronics['Date'].dt.month_name()
mysales_electronics
```

```python
# Showing overall prices
mysales_electronics['Price'] = mysales_electronics['Price'].astype(str).str.strip().astype(float)
mysales_electronics['Price']
```

```python
# Showing if there is any duplicates in the datatset
print(f"duplicates: {mysales_electronics.duplicated().sum()}")
```

```python
# Revenue generation list
mysales_electronics['Revennue'] = mysales_electronics['Price'] * mysales_electronics['Units']
mysales_electronics['Revennue']
```

```python
# pivot_table Graph facilitates the creation of summary tables that are both easy to read and analyze, aiding in pattern recognition, trend identification, and informed decision-making based on your data.
mysales_electronics.groupby('Date')['Revennue'].sum().plot()

```

```python
# this gives us for years of revenue in a plot graph
mysales_electronics.groupby('Year')['Revennue'].sum().plot()
```

```python
mysales_electronics.groupby('Sales Agent')['Revennue'].sum().plot()
```

```python
#  give us over performance of 
mysales_electronics.groupby('Branch')['Revennue'].sum().plot()
```

```python
# Gives us details of all sales agents and thier revenue generated
mysales_electronics.groupby('Sales Agent')['Revennue'].sum().nlargest(30)
```

```python
# Finally this functions give us over all electronic sales revenue of all products( Apple, compaq, Dell, Hp,Lenovo)

pivot = pd.pivot_table(mysales_electronics, values='Revennue', index="Products", columns="Year", aggfunc="sum")
pivot
```

```python
# Full Product Analysis 
mysales_electronics.groupby('Products')['Units'].sum().plot(kind='bar', color='blue', title=" Details Product Analysis")

```

```python

# Most Proftable Products
mysales_electronics.groupby('Products')['Revennue'].sum().plot(kind='bar', color='Red', title=" Most Performing Product Revenue Analysis")

```
